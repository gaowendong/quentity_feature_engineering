## 准备看看分段算法对于数据的表现状况

```python
import numpy as np
import os
import chardet
import pandas as pd
from pathlib import Path

import matplotlib.pylab as plt
import ruptures as rpt
import zigzag

from itertools import cycle
from ruptures.utils import pairwise
from matplotlib import gridspec


def price_zigzag(ax, price, pivots, changes):
    # plot price array data and zigzag
    ax.plot(np.arange(len(price)), price, 'k:', alpha=0.9)
    ax.plot(np.arange(len(price))[pivots != 0], price[pivots != 0], 'k-', linewidth=.5 )
    ax.scatter(np.arange(len(price))[pivots == 1], price[pivots == 1], color='r', s=5)
    ax.scatter(np.arange(len(price))[pivots == -1], price[pivots == -1], color='g', s=5)
    
    # color each (true) regime
    color_cycle = cycle(["#4286f4", "#f44174"])
    bkps = [0] + sorted(changes)
    alpha = 0.2  # transparency of the colored background
    for (start, end), col in zip(pairwise(bkps), color_cycle):
        ax.axvspan(max(0, start - 0.5), end - 0.5, facecolor=col, alpha=alpha)
    return ax

def volumn_changes(ax, volumn, changes):
    # plot price array data and zigzag
    ax.plot(np.arange(len(volumn)), volumn)
    
    # color each (true) regime
    color_cycle = cycle(["#00ff00", "#cc0000"])
    bkps = [0] + sorted(changes)
    alpha = 0.2  # transparency of the colored background
    for (start, end), col in zip(pairwise(bkps), color_cycle):
        ax.axvspan(max(0, start - 0.5), end - 0.5, facecolor=col, alpha=alpha)    
    return ax


stock_code = "000338"
filename = 'data/daydata/' + stock_code + ".csv"

start_index = -250
end_index = -170
with open(filename, 'rb') as f: 
    result = chardet.detect(f.read())

stock_data = pd.read_csv(filename, encoding=result['encoding'])
print(stock_data.iloc[start_index]['Date'] + " " + stock_data.iloc[end_index]['Date'])
# signal = stock_data[['Close','Volume']].to_numpy()[-270:]
signal_high = np.array(stock_data['High'])[start_index:end_index]
signal_low = np.array(stock_data['Low'])[start_index:end_index]
signal_volumn = np.array(stock_data['Volume'])[start_index:end_index]
# print(signal_volumn)
# print(signal[:2])
# print(signal[-2:])




#Changepoint detection with the Pelt search method
model="l2"
algo_high = rpt.Pelt(model=model).fit(signal_high)
result_high = algo_high.predict(pen=4)

model="l2"
algo_low = rpt.Pelt(model=model).fit(signal_low)
result_low = algo_low.predict(pen=4)

model="rbf"
algo = rpt.Pelt(model=model).fit(signal_volumn)
result_volumn = algo.predict(pen=3)


fig = plt.figure(figsize=(18, 12))
# fig, (ax1, ax2) = plt.subplots(nrows=2, ncols=1, sharex=True, figsize=(18, 12))
gs = gridspec.GridSpec(3, 1, height_ratios=[3, 3, 1])
ax1 = plt.subplot(gs[0])
ax2 = plt.subplot(gs[1], sharex = ax1)
ax3 = plt.subplot(gs[2], sharex = ax1)
# plt.setp(ax1.get_xticklabels(), visible=False)
plt.subplots_adjust(hspace=.0)

pivots = zigzag.peak_valley_pivots(signal_high, 0.01, -0.01)    
price_zigzag(ax1, signal_high, pivots, result_high) 

pivots = zigzag.peak_valley_pivots(signal_low, 0.01, -0.01) 
price_zigzag(ax1, signal_low, pivots, result_low)

pivots = zigzag.peak_valley_pivots(signal_low, 0.01, -0.01) 
price_zigzag(ax2, signal_low, pivots, result_low)

volumn_changes(ax3, signal_volumn, result_volumn)



```
